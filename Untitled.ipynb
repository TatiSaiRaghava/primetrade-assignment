```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(10, activation='relu', input_shape=(5,)),
    layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')
print("Model created successfully!")
```
